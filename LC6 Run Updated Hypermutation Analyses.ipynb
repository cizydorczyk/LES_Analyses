```bash
conda activate base
cd ~/les_complete/pi-testing/updated-hypermutator-testing/
pwd
```

```bash
# Filter recombinant SNPs from private genomes VCFs:
for i in $(cat ~/les_complete/les_isolate_list.txt); do filter_rc_vcfs.py --input_vcf ~/les_complete/snp_calling/L3-snp-calling/raw-snippy-output/${i}/snps.vcf --output_vcf rc-masked-VCFs/${i}.rc-masked.vcf --isolate ${i} --rc_regions ~/les_complete/snp_calling/L3-snp-calling/cfml/LES.rc_masked.without_outgroups.regions.txt --summary ~/les_complete/pi-testing/updated-hypermutator-testing/vcf-rc-filtering-summary.txt; done
```

```bash
# Filter recombinant SNPs from public genomes VCFs:
for i in $(cat ~/les_complete/public_genomes/public-les-list.txt); do filter_rc_vcfs.py --input_vcf ~/les_complete/snp_calling/L3-snp-calling/raw-snippy-output/${i}/snps.vcf --output_vcf rc-masked-VCFs/${i}.rc-masked.vcf --isolate ${i} --rc_regions ~/les_complete/snp_calling/L3-snp-calling/cfml/LES.rc_masked.without_outgroups.regions.txt --summary ~/les_complete/pi-testing/updated-hypermutator-testing/vcf-rc-filtering-summary.txt; done
```

'vcf-rc-filtering-summary.txt' contains one line per isolate with the # of SNPs in the original file (2nd column) and number remaining after filtering recombinant regions (3rd column).

```bash
# Run VCFtools TsTv for private genomes:
for i in $(cat ~/les_complete/les_isolate_list.txt); do vcftools --vcf ~/les_complete/pi-testing/updated-hypermutator-testing/rc-masked-VCFs/${i}.rc-masked.vcf --out ./tstv-ratios/${i} --TsTv-summary; done
```

```bash
# Run VCFtools TsTv for public genomes:
for i in $(cat ~/les_complete/public_genomes/public-les-list.txt); do vcftools --vcf ~/les_complete/pi-testing/updated-hypermutator-testing/rc-masked-VCFs/${i}.rc-masked.vcf --out ./tstv-ratios/${i} --TsTv-summary; done
```

```bash
# Parse private genomes with Python script:
for i in $(cat ~/les_complete/les_isolate_list.txt); do python ~/python_scripts/phd/i23_summarize_vcftools_tstv_1.py --input_log ./tstv-ratios/${i}.log --output_file ./tstv-summary.tsv --isolate ${i}; done
```
